```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1. Carica il dataset
df = pd.read_csv("drugs_dataset.tsv", sep="\t")

# Controlla le prime righe
print(df.head())

# 2. Crea un grafo non orientato (assunzione reciproca)
G = nx.Graph()

# 3. Aggiungi gli archi (ogni coppia di farmaci)
for _, row in df.iterrows():
    G.add_edge(row[0], row[1])

# 4. Controlla le dimensioni
print(f"Nodi: {G.number_of_nodes()}, Archi: {G.number_of_edges()}")
```

```python
# Calcolo centralit
centrality = nx.degree_centrality(G)
print(centrality)

# Layout con nodi pi distanti
pos = nx.spring_layout(G, k=1.0, scale=10, iterations=100)

plt.figure(figsize=(12, 10))
nx.draw(G, pos, with_labels=False, node_size=100, font_size=10)
plt.show()
```
