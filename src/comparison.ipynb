# Comparison: Italian Gang vs London Gang

## Between-network and sociocultural comparison (comparison between gangs)

### A. Structure and hierarchy
- **Comparison of density, clustering, modularity**
  - Assess differences in cohesion and internal organization.

- **Degree distribution**
  - Compare whether the network is more egalitarian or more centralized.

- **Hierarchy metrics (e.g., Gini coefficient on degree or centrality)**
  - Measures disparities of power/influence among members.


### B. Sociocultural patterns
- **Homophily**
  - **What it is:** Tendency to connect with similar members (nationality, age, role).
  - **Why useful:** Highlights cultural segregation or mixed groups.

- **Bridging / cross-cultural ties**
  - **Metrics:** Betweenness centrality of intercultural members.
  - **Why useful:** Identifies who mediates between ethnic or national subgroups.

- **Multilayer networks**
  - **What it is:** Considers different types of relationships (criminal, friendship, family, social media).
  - **Why useful:** Compares organization and cohesion across multiple levels; highlights different strategies.


### C. Comparative resilience and vulnerability
- **Targeted node removal**
  - Simulate the removal of leaders or brokers and observe the impact on cohesion and network diameter.

- **Assortativity**
  - Measures whether similar nodes tend to connect with each other.
  - Useful for understanding social structure and the inclusion/exclusion of new members.
